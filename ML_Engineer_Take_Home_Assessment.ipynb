# MODEL IMPLEMENTATION:

```python
import os
import torch
import torch.nn as nn
import numpy as np

from torchvision.models import resnet18, ResNet18_Weights
```

```python
pretrained_model = resnet18(weights=ResNet18_Weights.DEFAULT)
```

```python
class Model(nn.Module):

  def __init__(self):
    super(Model, self).__init__()
    pretrained_model = resnet18(weights=ResNet18_Weights.DEFAULT)
    self.backbone = nn.Sequential(*list(pretrained_model.children())[:-1])

    ### Initialize the required Layers
    self.have_object = nn.Linear(in_features=512, out_features=1)
    self.cat_or_dog = nn.Linear(in_features=512, out_features=3)
    self.specie = nn.Linear(in_features=512, out_features=10)
    self.bbox = nn.Linear(in_features=512, out_features=4)
    ### Initialize the required Layers

  def forward(self, input):
      out_backbone = self.backbone(input)[:,:,0,0]
      ### Write Forward Calls for the Model

      return {
          "bbox": nn.ReLU()(self.bbox(out_backbone)),
          "object": self.have_object(out_backbone),
          "cat_or_dog": self.cat_or_dog(out_backbone),
          "specie": self.specie(out_backbone)
      }
```

```python
specie_label_to_idx = {}
i = 0

for path in os.listdir("./assessment_dataset/labels/"):
  name = "_".join(path.split(".")[0].split("_")[:-1])
  if name not in specie_label_to_idx:
    specie_label_to_idx[name] = i
    i += 1

specie_label_to_idx['NA'] = len(specie_label_to_idx)

specie_idx_to_label = {val:key for (key, val) in specie_label_to_idx.items()}
```

```python
from bs4 import BeautifulSoup

def read_xml_file(path):
    with open(path, 'r') as f:
        data = f.read()
    bs_data = BeautifulSoup(data, 'xml')
    return {
        "cat_or_dog": bs_data.find("name").text,
        "xmin": int(bs_data.find("xmin").text),
        "ymin": int(bs_data.find("ymin").text),
        "xmax": int(bs_data.find("xmax").text),
        "ymax": int(bs_data.find("ymax").text),
        "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
    }
```

```python
from PIL import Image
```

```python
import torchvision.transforms as transforms
from torch.utils.data import Dataset

class CustomDataset(Dataset):

  def __init__(self, dataset_path, images_list, train=False):
    super(CustomDataset, self).__init__()
    self.preprocess = transforms.Compose([transforms.ToTensor(), transforms.Resize((224,224)), transforms.Normalize([0.485, 0.456, 0.406],[0.229, 0.224, 0.225])])
  
    image_folder_path = os.path.join(dataset_path, "images")
    label_folder_path = os.path.join(dataset_path, "labels")

    self.images = []
    self.labels = []

    for path in os.listdir(image_folder_path):
        name = path.split(os.sep)[-1].split(".")[0]
        if name in images_list:
          try:
            xml_path = os.path.join(label_folder_path, name+".xml")
            xml_data = read_xml_file(xml_path)
            self.labels.append(xml_data)
            self.images.append(Image.open(os.path.join(image_folder_path, path)))
          except:
            xml_data = {'cat_or_dog': 'NA',
                      'xmin': 0,
                      'ymin': 0,
                      'xmax': 0,
                      'ymax': 0,
                      'specie': 'NA'}
            self.labels.append(xml_data)
            self.images.append(Image.open(os.path.join(image_folder_path, path)))

  def __len__(self):
    assert len(self.images) == len(self.labels)

    return len(self.images)

  def __getitem__(self, index):
    image = self.preprocess(self.images[index])
    labels = self.labels[index]

    try:
      labels['specie'] = specie_label_to_idx[labels['specie']]
      labels['cat_or_dog'] = cat_or_dog_label_to_idx[labels['cat_or_dog']]
      labels['object'] = 1 if labels['cat_or_dog'] != 2 else 0
      labels['bbox'] = [labels['xmin'],labels['ymin'],labels['xmax'],labels['ymax']]
    except:
      pass
      
    return image, labels
```

```python
# import our library
import torchmetrics

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def train(epochs, model_weights):

  # Initialize Model and Optimizer
  model = Model()
  optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

  # Initialize Loss Functions
  have_object_loss = nn.BCEWithLogitsLoss()
  specie_loss = nn.CrossEntropyLoss()
  cat_or_dog_loss = nn.CrossEntropyLoss()
  #bbox_loss = nn.MSELoss(reduction='mean') # Not necessary you need to apply function to all coordinates together, You can have separete loss functions for all coordinates too
  # Below or Above
  xmin_loss = nn.MSELoss(reduction='mean')
  ymin_loss = nn.MSELoss(reduction='mean')
  xmax_loss = nn.MSELoss(reduction='mean')
  ymax_loss = nn.MSELoss(reduction='mean')

  training_dataset = CustomDataset("/content/assessment_dataset", images_list=train_list)
  training_loader = train_dataloader

  if torch.cuda.is_available():
    model = model.cuda()

  def train_one_epoch(epoch_index, tb_writer):
      running_loss = 0.
      last_loss = 0.

      # Here, we use enumerate(training_loader) instead of
      # iter(training_loader) so that we can track the batch
      # index and do some intra-epoch reporting
      for i, data in enumerate(training_loader):
          # Every data instance is an input + label pair
          inputs, labels = data
          if torch.cuda.is_available():
            inputs = inputs.cuda()

            #for key in labels:
            #  labels[key] = labels[key].cuda()

          # Make predictions for this batch
          outputs = model(inputs)

          # Compute the loss and its gradients
          loss_have_object = have_object_loss(outputs['object'].to(torch.float32),labels['object'].to(device).unsqueeze(1).float().to(torch.float32)) #have_object_loss(outputs["object"], None)
          loss_specie = have_object_loss(outputs['cat_or_dog'].to(torch.float32),torch.nn.functional.one_hot(labels['cat_or_dog'],len(cat_or_dog_idx_to_label)).float().to(device)) #specie_loss(outputs["specie"], None)
          loss_cat_or_dog = have_object_loss(outputs['cat_or_dog'].to(torch.float32),torch.nn.functional.one_hot(labels['cat_or_dog'],len(cat_or_dog_idx_to_label)).float().to(device)) #cat_or_dog_loss(outputs["cat_or_dog"], None)
          
          #loss_bbox = bbox_loss(outputs["bbox"], None)
          # Above or Below
          loss_xmin = xmin_loss(outputs["bbox"][:,0].to(torch.float32), labels['xmin'].to(device).to(torch.float32))
          loss_ymin = ymin_loss(outputs["bbox"][:,1].to(torch.float32), labels['ymin'].to(device).to(torch.float32))
          loss_xmax = xmax_loss(outputs["bbox"][:,2].to(torch.float32), labels['xmax'].to(device).to(torch.float32))
          loss_ymax = ymax_loss(outputs["bbox"][:,3].to(torch.float32), labels['ymax'].to(device).to(torch.float32))

          loss_bbox = 0.25*(loss_xmin+loss_ymin+loss_xmax+loss_ymax)

          # Assign more weights to species
          loss = 10*loss_have_object +  100*loss_specie + 10*loss_cat_or_dog + 0.0001 * loss_bbox # Consolidate all individual losses

          #print (loss)

          optimizer.zero_grad()
          loss.backward()
          optimizer.step()

          # Gather data and report
          running_loss += loss.item()
          if i % 10 == 0:
              last_loss = running_loss / 10 # loss per batch
              running_loss = 0.
      return last_loss

  last_metric = 0

  for i in range(epochs):
    model.train()
    epoch_loss = train_one_epoch(i, None)
    print(f' Epoch {i} Loss : {epoch_loss}')

    model.eval()
    metrics = test(model, val_list)
    print(metrics)

    # Save model weights if it improves specie accuracy
    if metrics[2] >= last_metric:
      last_metric = metrics[2]
      torch.save(model.state_dict(), "model.pth")

def test(model, val_list):

  from sklearn.metrics import mean_squared_error, accuracy_score

  val_dataset = CustomDataset(dataset_path='./assessment_dataset/', images_list=val_list)
  val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)

  metric_object = accuracy_score
  metric_cat_or_dog = accuracy_score
  metric_specie = accuracy_score
  metric_bbox = mean_squared_error

  act_object = []
  act_cat_or_dog = []
  act_specie = []
  act_xmin = []
  act_ymin = []
  act_xmax = []
  act_ymax = []
  act_bbox = []

  pred_object = []
  pred_cat_or_dog = []
  pred_specie = []
  pred_xmin = []
  pred_ymin = []
  pred_xmax = []
  pred_ymax = []
  pred_bbox = []

  for i, data in enumerate(val_loader):
      inputs, labels = data

      if torch.cuda.is_available():
        inputs = inputs.cuda()

      # Make predictions for this batch
      outputs = model(inputs)

      act_object += labels['object'].detach().cpu().numpy().tolist()
      act_cat_or_dog += labels['cat_or_dog'].detach().cpu().numpy().tolist()
      act_specie += labels['specie'].detach().cpu().numpy().tolist()
      act_xmin += labels['xmin'].detach().cpu().numpy().tolist()
      act_ymin += labels['ymin'].detach().cpu().numpy().tolist()
      act_xmax += labels['xmax'].detach().cpu().numpy().tolist()
      act_ymax += labels['ymax'].detach().cpu().numpy().tolist()
      #act_bbox.append(np.asarray(labels['bbox']))

      pred_object += nn.Sigmoid()(outputs['object'])[:,0].detach().cpu().numpy().round().astype(int).tolist()
      pred_cat_or_dog += nn.Softmax(-1)(outputs['cat_or_dog']).detach().cpu().numpy().argmax(-1).tolist()
      pred_specie += nn.Softmax(-1)(outputs['specie']).detach().cpu().numpy().argmax(-1).tolist()
      #pred_xmin.append(outputs['bbox'][0].detach().cpu().numpy().tolist())
      #pred_ymin.append(outputs['bbox'][1].detach().cpu().numpy().tolist())
      #pred_xmax.append(outputs['bbox'][2].detach().cpu().numpy().tolist())
      #pred_ymax.append(outputs['bbox'][3].detach().cpu().numpy().tolist())
      pred_bbox.append(outputs['bbox'].detach().cpu().numpy())

  #act_bbox = np.concatenate(act_bbox,0)
  pred_bbox = np.concatenate(pred_bbox,0)

  score_object = metric_object(act_object, pred_object)
  score_cat_or_dog = metric_cat_or_dog(act_cat_or_dog, pred_cat_or_dog)
  score_specie = metric_specie(act_specie, pred_specie)
  score_bbox = 0.25*(metric_bbox(act_xmin, pred_bbox[:,0]) + metric_bbox(act_ymin, pred_bbox[:,1]) + \
                     metric_bbox(act_xmax, pred_bbox[:,2]) + metric_bbox(act_ymax, pred_bbox[:,3]))

  return score_object, score_cat_or_dog, score_specie, score_bbox
```

```python
from PIL import ImageDraw

def visualize(model_weights, image_name, output_folder="output"):

  model = Model()
  model.load_state_dict(torch.load(model_weights))

  try:
    image = np.asarray(Image.open(os.path.join("/content/assessment_dataset/images", image_name+".jpg")))
  except:
    image = np.asarray(Image.open(os.path.join("/content/assessment_dataset/images", image_name+".jpeg")))

  preprocess = transforms.Compose([transforms.ToTensor(), transforms.Resize((224,224)), transforms.Normalize([0.485, 0.456, 0.406],[0.229, 0.224, 0.225])])
  model.eval()

  with torch.no_grad():
    output = model(preprocess(image).unsqueeze(0))
  
  pred_object = nn.Sigmoid()(output['object'])[:,0].detach().cpu().numpy().round().astype(int).tolist()[0]
  pred_cat_or_dog = nn.Softmax(-1)(output['cat_or_dog']).detach().cpu().numpy().argmax(-1).tolist()[0]
  pred_specie = nn.Softmax(-1)(output['specie']).detach().cpu().numpy().argmax(-1).tolist()[0]
  pred_bbox = output['bbox'].detach().cpu().numpy()[0,:]

  if pred_object == 1:
    return {"has_object": True, "cat_or_dog": cat_or_dog_idx_to_label[pred_cat_or_dog], \
          "specie": specie_idx_to_label[pred_specie], "xmin": int(pred_bbox[0]), "ymin": int(pred_bbox[1]), \
          "xmax": int(pred_bbox[2]), "ymax": int(pred_bbox[3])}
  else:
    return {"has_object": False, "cat_or_dog": "NA", "specie": "NA", "xmin": 0, "ymin": 0, "xmax": 0, "ymax": 0
}
```
